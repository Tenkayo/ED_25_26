````python
data_spam = {
    'Słowo_1': ['darmowy', 'darmowy', 'spotkanie', 'raport', 'oferta', 'darmowy',
                'spotkanie', 'oferta', 'raport', 'darmowy', 'spotkanie', 'oferta'],
    'Słowo_2': ['wygrana', 'wygrana', 'jutro', 'kwartalny', 'specjalna', 'rabat',
                'dziś', 'limitowana', 'miesięczny', 'rabat', 'pilne', 'wyjątkowa'],
    'Wykrzyknik': ['TAK', 'TAK', 'NIE', 'NIE', 'TAK', 'TAK',
                   'NIE', 'TAK', 'NIE', 'TAK', 'NIE', 'TAK'],
    'Spam': ['TAK', 'TAK', 'NIE', 'NIE', 'NIE', 'TAK',
             'NIE', 'NIE', 'NIE', 'TAK', 'NIE', 'NIE']
}

df = pd.DataFrame(data_spam)
print(df)

class_counts = df['Spam'].value_counts()
apriori = class_counts / len(df)
print(f"P(SPAM = TAK): {apriori['TAK']}")
print(f"P(SPAM = NIE): {apriori['NIE']}")
prawdopodobienstwa = {}

for kolumna in ['Słowo_1', 'Słowo_2', 'Wykrzyknik']:
    print(f"\nPrawdopodobieństwa warunkowe dla kolumny '{kolumna}':")
    prawdopodobienstwa[kolumna] = {}
    for klasa in class_counts.index:
        warunki = df[df['Spam'] == klasa][kolumna].value_counts()
        print(f"\nKlasa: {klasa}")
        prawdopodobienstwa[kolumna][klasa] = {}
        for wartosc in warunki.index:
          prawdopodobienstwa[kolumna][klasa][wartosc] = warunki[wartosc]/class_counts[klasa]
          print(f"P({wartosc} | {klasa}) = {warunki[wartosc]}/{class_counts[klasa]} = {prawdopodobienstwa[kolumna][klasa][wartosc]}")

nowy_mail = {
    'Słowo_1': 'darmowy',
    'Słowo_2':  'wygrana',
    'Wykrzyknik': 'TAK'
}

for klasa in ['TAK','NIE']:
  print(f"\nDla klasy: {klasa}")
  predykcja = apriori[klasa]
  print(f"Prior: {predykcja}")
  for key, val in nowy_mail.items():
    modyfikator = prawdopodobienstwa[key][klasa].get(val, 0)
    print(f"P({key} = {val} | {klasa}) = {modyfikator}")
    predykcja *= modyfikator
    print(f"Wynik: {predykcja}")
````
